```python
def barrier(z): return 0.1 - z *np.conj(z)
trange = np.arange(0, 2*np.pi, 0.01)

for i in range(25):
    ang = np.exp(1j* randang())
    r = np.random.uniform(np.sqrt(0.9),1)
    # r = np.sqrt(0.1)
    x_0 = np.array([r, np.sqrt(1-r**2) * ang])

    xs = [np.dot(had_ham(t), x_0) for t in trange]
    z = [x[0] for x in xs]
    b = [barrier(ele) for ele in z]

    plt.plot(trange, b)
    plt.ylim(-1, 0.2)

plt.axhline(y=0, c="k")
plt.plot(trange, [np.sin(2*t - 1)/3 - 1/2 for t in trange], '-.')
plt.show()
```
